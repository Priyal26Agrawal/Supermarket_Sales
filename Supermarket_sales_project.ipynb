# Data Analysis

```python
import datetime
supermarket_sales["sales_month"]=supermarket_sales["Date"].dt.month_name()
supermarket_sales
```

```python
supermarket_sales.columns
```

```python
#Product line wise total_sales,total tax_5% and average rating
product_line_sales_tax=supermarket_sales.groupby(["Product line"]).agg({"Total":"sum","Tax 5%":"sum","Rating":"mean"}).reset_index()
product_line_sales_tax
```

```python
product_line_sales_tax=product_line_sales_tax.rename({"Total":"Total_sales(including tax)","Tax 5%":"Total_tax(5%)","Rating":"Avg_rating"},axis=1)
product_line_sales_tax=product_line_sales_tax.round(2)
product_line_sales_tax
```

```python
#Customer type wise total_sales,total tax_5% and average rating
customer_type_sales_tax=supermarket_sales.groupby(["Customer type"]).agg({"Total":"sum","Tax 5%":"sum","Rating":"mean"}).reset_index()
customer_type_sales_tax=customer_type_sales_tax.rename({"Total":"Total_sales(including tax)","Tax 5%":"Total_tax(5%)","Rating":"Avg_rating"},axis=1)
customer_type_sales_tax=customer_type_sales_tax.round(2)
customer_type_sales_tax
```

```python
#Product line ,gender wise total unit price
Product_line_gender_price= supermarket_sales.groupby(["Product line","Gender"]).agg({"Unit price":"sum"}).reset_index()
Product_line_gender_price
```

```python
#Product line wise average unit price
Product_line_price= supermarket_sales.groupby(["Product line"]).agg({"Unit price":"mean"}).reset_index()
Product_line_price
```

```python
#Month wise total quantity sold and total sale
monthly_sale= supermarket_sales.groupby(["sales_month"]).agg({"Quantity":"sum","Total":"sum"}).reset_index()
monthly_sale=monthly_sale.round(2)
monthly_sale=monthly_sale.sort_values(by="Quantity",ascending=False)
monthly_sale
```

```python
#Payment method wise Gross Income
Payment_method_income=supermarket_sales.groupby(["Payment"]).agg({"gross income":"sum","Rating":"mean"}).reset_index()
Payment_method_income=Payment_method_income.sort_values(by="gross income",ascending= False).round(2)
Payment_method_income
```
